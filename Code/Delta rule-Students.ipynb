```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
np.random.seed(42)

N = 100

xs = []

for _ in range(N):
    x0 = np.random.randn()-1
    x1 = np.random.randn()-1
    xs.append((x0, x1))
    
    x0 = np.random.randn()+2
    x1 = np.random.randn()+1
    xs.append((x0, x1))
    
xs = np.asarray(xs)
print(xs.shape)

plt.scatter(*xs[::2].T, c='r')
plt.scatter(*xs[1::2].T, c='b')
plt.axhline(0)
plt.axvline(0)
plt.ylim([-5,5])
plt.xlim([-5,5])
```

```python
w = np.random.randn(2)
eta = 0.01
x = np.linspace(-5,5)
ys = []

for i in range(2*N):
    # Determine class/target
    t = -1 if i % 2 == 0 else 1

    # Compute prediction
    pred = w @ xs[i]

    # Update weights
    w = w + eta * (t-pred) * xs[i]

    # Plot decision boundary
    y = -w[0]/w[1] * x
    ys.append(y)
    plt.plot(x,y, 'k-', alpha=.02)


# Plot final
plt.scatter(*xs[::2].T, c='r')
plt.scatter(*xs[1::2].T, c='b')
plt.plot(x,y, 'k--')
plt.ylim([-5,5])
plt.xlim([-5,5])
```
